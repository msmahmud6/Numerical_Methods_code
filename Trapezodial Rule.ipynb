```python
def f(x):
  return 4*x**4+3*x**2+2*x+5
```

```python
lower_limit = 2
upper_limit = 6
step_value = 0.25

x_value = []
fx_value = []

i=2
while i<=6:
  x_value.append(i)
  fx_value.append(f(i))
  i += step_value

print(x_value)
print(fx_value)
```

```python
def trapezodial(y_value, h):
  y0 = float(y_value[0])
  yn = y_value[len(y_value)-1]
  y1_range_sum = sum(y_value[1:len(y_value)-1])
  result = (h/2) * (y0 + yn + 2*y1_range_sum)
  return result
```

```python
print(trapezodial(fx_value,step_value))
```
