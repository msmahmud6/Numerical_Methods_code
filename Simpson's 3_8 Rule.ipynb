```python
def f(x):
  return 4*x**4+3*x**2+2*x+5
```

```python
lower_limit = 2
upper_limit = 6
step_value = 0.25

x_value = []
fx_value = []

i=lower_limit
while i<=upper_limit:
  x_value.append(i)
  fx_value.append(f(i))
  i += step_value

print(x_value)
print(fx_value)
```

```python
def simpsons_3_8(y_value, h):
  y0 = float(y_value[0])
  yn = y_value[len(y_value)-1]
  threes_multiple_sum = sum(y_value[i] for i in range(len(y_value)) if i % 3 == 0)
  rest_sum = sum(y_value[i] for i in range(len(y_value)) if i % 3 != 0)
  result = (3*h/8) * (y0 + yn + 2*threes_multiple_sum + 3*rest_sum)
  return result
```

```python
simpsons_3_8(fx_value, step_value)
```
