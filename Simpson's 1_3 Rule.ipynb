```python
def f(x):
  return 4*x**4+3*x**2+2*x+5
```

```python
lower_limit = 2
upper_limit = 6
step_value = 0.25

x_value = []
fx_value = []

i=lower_limit
while i<=upper_limit:
  x_value.append(i)
  fx_value.append(f(i))
  i += step_value

print(x_value)
print(fx_value)
```

```python
def simpsons_1_3(y_value, h):
  y0 = float(y_value[0])
  yn = y_value[len(y_value)-1]
  odd_range_sum = sum(y_value[1:len(y_value)-1:2])
  even_range_sum = sum(y_value[2:len(y_value)-1:2])
  result = (h/3) * (y0 + yn + 4*odd_range_sum + 2*even_range_sum)
  return result
```

```python
simpsons_1_3(fx_value, step_value)
```
